```python
import os
import pandas as pd

# Define the directory where the CSV files are located
directory = '/Users/zhanglu/Downloads/25FebDownloadedBackup/'

# Initialize an empty list to store the DataFrames
dfs = []

# Loop through all files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        # Construct the full file path
        filepath = os.path.join(directory, filename)
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(filepath)
        
        # Append the DataFrame to the list
        dfs.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Now combined_df contains the data from all CSV files in the directory
print(combined_df)
```

```python
# count every month's 评论星级's amount
monthly_counts = new_combined_df.groupby(['评论时间', '评论星级']).size().unstack(fill_value=0)
monthly_counts
```

```python
# Plot the data
monthly_counts.plot(kind='bar', stacked=True)

# Set the labels and title
plt.xlabel('评论月份')
plt.ylabel('评论数量')
plt.title('每月不同评论星级的评论数量')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(50, 30))  # Setting the size of the plot

for i in range(1, 6):  # Looping over each star rating
    plt.scatter(monthly_counts['评论时间'], monthly_counts[f'评论星级{i}'], label=f'评论星级{i}')

plt.xlabel('comment time')
plt.ylabel('comment amount')
plt.title('amount of different star rating comments by month')
plt.legend()
plt.grid(True)
plt.show()
```

```python
monthly_counts['总评论数量'] = monthly_counts.iloc[:, 1:].sum(axis=1)

# Calculate the ratio of each star rating's comments
for i in range(1, 6):
    monthly_counts[f'评论星级{i}_比例'] = monthly_counts[f'评论星级{i}'] / monthly_counts['总评论数量']

# Plotting each star rating's count ratio against time
plt.figure(figsize=(10, 6))  # Setting the size of the plot

for i in range(1, 6):  # Looping over each star rating
    plt.scatter(monthly_counts['评论时间'], monthly_counts[f'评论星级{i}_比例'], label=f'评论星级{i}')

plt.xlabel('评论时间')
plt.ylabel('比例')
plt.title('每月不同评论星级的评论比例')
plt.legend()
plt.grid(True)
plt.show()
```
